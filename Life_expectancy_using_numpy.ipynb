<h1>Now we have converted both our Xs and ys into numpy arrays.</h1>

```python
def calculate_cost(y_actual,y_predicted):
    num=y_actual.shape[0]
    cost=(1/(2*num))*np.sum(np.square(y_actual-y_predicted))
    return cost
```

```python
def hypothesis(X,theta):
    y_predicted=np.dot(X,theta)
    return y_predicted
```

<h1>Applying Gradient Descent.</h1>

```python
def gradient_descent(X,y_actual,alpha,num_iterations):
    num=y_actual.shape[0]
    theta=np.zeros(X.shape[1])
    
    for i in range(num_iterations):
        predictions=hypothesis(X,theta)
        error=predictions-y_actual
        gradient=(1/num)*np.dot(X.T,error)
        theta-=alpha*gradient

    return calculate_cost(y_actual,hypothesis(X,theta))
        
```

<h2>Calculating the accuracy.</h2>

```python
mse=gradient_descent(X_df_numpy,y_df_numpy,0.01,10000)
np.sqrt(mse)
print("The Mean Square Error is: ",mse)
```
